```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np


def openimg(path):
    img = mpimg.imread(path)
    return img

def gray(img):
    y=[]
    for z in range(len(img)):
        x=[]
        for s in range(len(img[z])):
            x.append((img[z,s,0]+img[z,s,1]+img[z,s,2])/3)
        y.append(x)    
    return y

def inverse(img):
    for z in range(len(img)):
        for s in range(len(img[z])):
            img[z,s]=255-img[z,s]
    return img






path = input('Path:')
bsp = openimg(path+'.png')
plt.axis('off')
plt.imshow(bsp)
plt.show()


bspgray=gray(bsp)
plt.axis('off')
plt.gray()
plt.imshow(bspgray)
plt.savefig(path+'_gray.png')
plt.show()




```
