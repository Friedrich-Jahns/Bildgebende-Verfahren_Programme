```python
from matplotlib import pyplot as plt
import numpy as np
from skimage import color
from skimage import io
```

```python
kernelsize = 501

if kernelsize%2==1:
    ks=kernelsize
ksn=int((ks-1)/2)
```

```python
#Get RGB_Channels from 4 Channel png
imgi = io.imread('test.png')[:,:,:3]

#convert to Grayscale
img = color.rgb2gray(imgi)
plt.imshow(img)
plt.gray()
plt.axis('off')
plt.show()
```

```python
#   1ol 2o  1or
#   3l  org   3r
#   1ul 2u  1ur

m2 = np.flipud(img)
m2o = m2[-ksn:,:]
m2u = m2[:ksn,:]

m1 = np.fliplr(m2)
m1ol=m1[-ksn:,-ksn:]
m1ul=m1[:ksn,-ksn:]
m1ur=m1[:ksn,:ksn]
m1or=m1[-ksn:,:ksn]

m3 = np.fliplr(img)
m3l= m3[:,-ksn:]
m3r= m3[:,:ksn]
```

```python
imgn=np.zeros((len(img)+2*ksn,len(img[0])+2*ksn))
imgn[0:ksn,0:ksn]=m1ol
imgn[ksn:-ksn,0:ksn]=m3l
imgn[-ksn:,:ksn]=m1ul
imgn[-ksn:,ksn:-ksn]=m2u
imgn[-ksn:,-ksn:]=m1ur
imgn[ksn:-ksn,-ksn:]=m3r
imgn[0:ksn,-ksn:]=m1or
imgn[0:ksn,ksn:-ksn]=m2o
imgn[ksn:-ksn,ksn:-ksn]=img
```

```python
plt.imshow(imgn)
plt.axis('off')
plt.show()
```
