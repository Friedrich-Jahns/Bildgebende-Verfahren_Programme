```python
from scipy import misc
from matplotlib import pyplot as plt
import numpy as np




def blackwhite(img):
    for z in range(len(img)):
        for s in range(len(img[z])):
            
            if img[z,s]<128:
                img[z,s]=0
            else:
                img[z,s]=255
    return img


def transpose(img):
    return img.T


def subtract(img1,img2):
    return img1-img2

def rescale(img):
    img= img-np.min(img)
    img= 255*(img/max(img))
    return img

ascent = misc.ascent()
print(ascent)
plt.gray()
plt.imshow(ascent)
plt.axis('off')
#rescale(ascent)
plt.show()





nascent=subtract(ascent,transpose(ascent))
    
    

print(nascent)
plt.axis('off')
plt.gray()
plt.imshow(nascent)
#rescale(nascent)
plt.show()
```
